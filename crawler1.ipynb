```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

titles_text=[]
result={}

def crawler():    

    page = 1  
    list = ['0006110000','0006120000','0006130000','0006140000','0006150000',
            '6000069377','6000061257','6000023665','6000023666','6000023669','6000025610','6000014059']
    endPages = [90,113,54,121,74,90,27,129,81,133,141,172]
    
    for j in range(len(list)):
        
        print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
        page = 1
        titles_text=[]
        
        while page <= endPages[j]: 
            url = "http://emart.ssg.com/category/main.ssg?dispCtgId=" + list[j] + "&page=" + str(page)
            print(url)
            response = requests.get(url)
            html = response.content

            soup = BeautifulSoup(html, 'html.parser')

            titles_html = soup.select('li > div > div > div > a >.tx_ko')
            
            for i in range(len(titles_html)):
                titles_text.append(titles_html[i].text)
            
            result= {"title":titles_text}  

            df = pd.DataFrame(result)
            df1 = df.drop_duplicates()
            page += 1
            
        df1.to_csv('7.csv', encoding = 'cp949',header = False, mode='a')   

def main():
    crawler() 
    
main()
```
